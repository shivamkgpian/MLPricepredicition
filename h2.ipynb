```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # statistical data visualization
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the csv file as data
df1=pd.read_csv("train.csv")
df2=pd.read_csv("test.csv")
# you may print first few rows
# data.head()
df1.head()
```

```python
def fill_maincateg(df):
    for ind, item in enumerate(df.maincateg):
        # print(item)
        
        # how else to check if item is nan
        if(item!="Men" and item != "Women"):
            # print(df.title[ind])
            #if(df.title[ind].str.contains('Men')):
            if("Men" in df.title[ind]):
                df.loc[ind, "maincateg"] = 'Men'
            else:
                df.loc[ind, "maincateg"] = 'Women'
    print("Done")
    
    return df
```

```python
fill_maincateg(df1)
fill_maincateg(df2)
```

```python
df1['star_5f'] = df1['star_5f'].fillna(df1['star_5f'].mean())
df1['star_4f'] = df1['star_4f'].fillna(df1['star_4f'].mean())
df1['star_3f'] = df1['star_3f'].fillna(df1['star_3f'].mean())
# df1['star_5f'] = df1['star_5f'].fillna(df1['star_5f'].median())
# df1['star_4f'] = df1['star_4f'].fillna(df1['star_4f'].median())
# df1['star_3f'] = df1['star_3f'].fillna(df1['star_3f'].median())
df1['noreviews1'] = df1['noreviews1'].fillna(df1['noreviews1'].mean())
df1['maincateg'].replace({'Men':1, 'Women':0}, inplace=True)
df1['platform'].replace({'Flipkart':1, 'Amazon':0}, inplace=True)
```

```python
df1.isnull().sum()
```

```python
#df1 = df1.dropna(how='any',axis=0)
# df2 = df2.dropna(how='any',axis=0)
coloumn=['star_5f','star_4f','star_3f','star_2f','star_1f']
df1['norating1'] = df1[coloumn].sum(axis=1)
```

```python
#df2['Rating'] = df2['Rating'].fillna(df2['Rating'].mean())
df2['maincateg'].replace({'Men':1, 'Women':0}, inplace=True)
df2['platform'].replace({'Flipkart':1, 'Amazon':0}, inplace=True)
#df2['star_5f'] = df2['star_5f'].fillna(df2['star_5f'].mean())
#df2['star_1f'] = df2['star_1f'].fillna(df2['star_1f'].mean())
```

```python
cnt=0
for i in range(len(df2)):
    if(pd.isna(df2['star_5f'][i])&pd.isna(df2['star_1f'][i])):
        cnt+=1
    elif(pd.isna(df2['star_5f'][i])):
        df2['star_5f'][i]=df2['norating1'][i]-(df2['star_1f'][i]+df2['star_2f'][i]+df2['star_3f'][i]+df2['star_4f'][i])
    elif(pd.isna(df2['star_1f'][i])):
        df2['star_1f'][i]=df2['norating1'][i]-(df2['star_5f'][i]+df2['star_2f'][i]+df2['star_3f'][i]+df2['star_4f'][i])
    else:
        continue
cnt
```

```python
for i in range(len(df2)):
    if(pd.isna(df2['star_5f'][i])&pd.isna(df2['star_1f'][i])):
        df2['star_5f'][i]=round((df2['norating1'][i]-df2['star_2f'][i]-df2['star_3f'][i]-df2['star_4f'][i])/2)
        df2['star_1f'][i]=df2['norating1'][i]-df2['star_2f'][i]-df2['star_3f'][i]-df2['star_4f'][i]-df2['star_5f'][i]
```

```python
for i in range(len(df2)):
    if(pd.isna(df2['Rating'][i])):
        if(df2['norating1'][i]==0):
            df2['Rating'][i]=0
        else:
            df2['Rating'][i]=(df2['star_5f'][i]*5+df2['star_4f'][i]*4+df2['star_3f'][i]*3+df2['star_2f'][i]*2+df2['star_1f'][i]*1)/df2['norating1'][i]
```

```python
X_train = df1.drop(['id','Offer %','price1','title','star_5f','star_4f','star_3f','star_2f','star_1f'], axis=1)

y_train = df1['price1']
```

```python
X_test=df2.drop(['id','title','star_5f','star_4f','star_3f','star_2f','star_1f'], axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor,StackingRegressor
import xgboost as xg
from sklearn.linear_model import LinearRegression
```

```python
classifier=RandomForestRegressor(max_depth=30,max_features=5,min_samples_leaf=1,min_samples_split=2,n_estimators=580,bootstrap=True)  
classifier.fit(X_train, y_train)
y1_pred=classifier.predict(X_test)
```

```python
xgb_r = xg.XGBRegressor(max_depth=6,n_estimators =700,learning_rate=0.4,gamma=0.7,eval_metric='rmse')
xgb_r.fit(X_train, y_train)
y2_pred=xgb_r.predict(X_test)
```

```python
y_pred=[None]*len(y2_pred)
```

```python
for i in range(len(y_pred)):
    y_pred[i]=2*y1_pred[i]*y2_pred[i]/(y1_pred[i]+y2_pred[i])
```

```python
models=[('classifier',RandomForestRegressor(max_depth=30,max_features=5,min_samples_leaf=1,min_samples_split=2,n_estimators=580,bootstrap=True)),
        ('xgb_r', xg.XGBRegressor(max_depth=6,n_estimators =700,learning_rate=0.4,gamma=0.7,eval_metric='rmse'))]
```

```python
meta= LinearRegression()
```

```python
stack=StackingRegressor(estimators=models,final_estimator=meta,cv=5)
stack.fit(X_train, y_train)
y3_pred=stack.predict(X_test)
```

```python
for i in range(len(y_pred)):
    y_pred[i]=(y_pred[i]+y3_pred[i])/2
```

```python
i_d=df2['id']
result_1 = pd.DataFrame(data=i_d, columns=['id'])
result_1['price1']=y_pred
result_1.to_csv("pred_Result_14.csv", index = False)
```
